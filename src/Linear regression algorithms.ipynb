# Linear regression

n - number of features <br>
m - number of training examples <br>
$x^{(i)}$ - $i^{th}$ input variable <br>
$y^{(i)}$ - $i^{th}$ output variable

## Hypothesis

Parameters: <br>
&emsp; $\theta_0, \theta_1,...\theta_n = \theta$ <br><br>
Hypothesis: <br>
&emsp; $h_\theta(x) = \theta_0x_0+\theta_1x_1+...+\theta_nx_n = \theta^Tx$ &emsp; for $x_0=1$<br><br>
Vectorized hypothesis: <br>
&emsp; $y_{predicted} = X\theta$

## Cost function - Mean squared error

Cost function: <br>
&emsp; $J(\theta) = \frac{1}{2m} \sum \limits _{i=1} ^m (h_\theta(x^{(i)})-y^{(i)})^2$ <br><br>
Vectorized cost function: <br>
&emsp; $J(\theta) = \frac{1}{2m}(X\theta - y)^T(X\theta - y)$

## Gradient descent

$\alpha$ - learning rate <br><br>
Gradient descent algorithm: <br>
&emsp; $\theta_j := \theta_j - \alpha \frac{\partial}{\partial\theta_j}J(\theta)$ <br>
&emsp;$\theta_j := \theta_j - \alpha \frac{1}{m} \sum \limits _{i=1} ^m (h_\theta(x^{(i)})-y^{(i)})x_j^{(i)}$ &emsp; for j=0...n,  $x_0=1$ <br><br>
Vectorized gradient descent: <br>
&emsp; $\theta = \theta - \alpha \frac{1}{m} ((X\theta - y)^T X)^T = \theta - \alpha \frac{1}{m} X^T (X\theta - y)$

## Feature scaling and mean normalization

$\mu_i - mean$ <br>
$\sigma_i$ - standard deviation <br><br>
Feature scaling and mean normalization: <br>
&emsp; $x_i = \frac{x_i - \mu_i}{\sigma_i}$ <br><br>
Vectorized feature scaling and mean normalization: <br>
&emsp; $\mu \in \mathbb{R^n}$ <br>
&emsp; $\sigma \in \mathbb{R^n}$ <br>
&emsp; $t \in \mathbb{1}^m$ <br><br>
&emsp; $X_{normalize} = (X - t\mu^T) \oslash (t\sigma^T)$

## Normal equation

min of $J(\theta) = \frac{\partial}{\partial\theta}J(\theta) = \frac{\partial}{\partial\theta} \frac{1}{2m} \sum \limits _{i=1} ^m (h_\theta(x^{(i)})-y^{(i)})^2 = 0 => \theta = (X^TX)^{-1}X^Ty$ <br>

## Learning rate decay

$\eta = \eta * \frac{1}{1 + decay\_rate * epoch}$

# Linear regression implementations

## Mean squared error implementation

```python
def mean_squared_error(X, y, theta):
    m = X.shape[0]
    J = (1/(2*m)) * ((X.dot(theta)-y).T).dot(X.dot(theta)-y)
    return J[0][0]
```

## Gradient descent implementation

```python
def gradient_descent(X, y, theta, alpha):
    m = X.shape[0]
    return theta - (alpha/m) * X.T.dot(X.dot(theta)-y)
```

## Feature normalization implementation

```python
def feature_normalize(X):
    mu = np.array([np.mean(X, axis=0)])
    sigma = np.array([np.std(X, axis=0)])
    t = np.ones((X.shape[0], 1))
    X_norm = (X - t.dot(mu)) / (t.dot(sigma))
    return X_norm, mu, sigma
```

## Normal equation implementation

```python
def normal_equation(X, y):
    return np.linalg.inv(X.T.dot(X)).dot(X.T.dot(y))
```

## Gradient descent with learning rate decay

```python
def decay_learning_rate(eta, decay_rate, epoch):
    return eta*(1/(1+decay_rate*epoch))
```

## Plot hypothesis

```python
import numpy as np
from matplotlib import pyplot as plt

# Columns: population of city, profit
data = np.loadtxt('datasets/food_truck.csv', delimiter=',')

X = data[:, 0:1]
y = data[:, 1:2]

plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.plot(X, y, 'xr')

theta = np.array([[-1], [2]])
min_max = np.array([[np.min(X)], [np.max(X)]])
min_max_with_x0 = np.hstack((np.ones((2,1)), min_max))
pred = min_max_with_x0.dot(theta) 

plt.plot(min_max, pred)
plt.show()
```

## Plot different hypothesis and mean squared error values

```python
import numpy as np
from matplotlib import pyplot as plt

data = np.loadtxt('datasets/food_truck.csv', delimiter=',')

m = data.shape[0]
X = np.hstack((np.ones((m, 1)), data[:, 0:1]))
y = data[:, 1:2]

min_max = np.array([[np.min(X[:, 1])], [np.max(X[:, 1])]])
min_max_with_x0 = np.hstack((np.ones((2,1)), min_max))

plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.plot(X[:, 1:2], y, 'xr')

thetas = np.array([
    [[-1], [2]],
    [[-4], [1]],
    [[-13], [1]]
])

for theta in thetas:
    J = mean_squared_error(X, y, theta)
    pred = min_max_with_x0.dot(theta)
    label = f"J={np.around(J, 2)}"
    plt.plot(min_max, pred, label=label)

plt.legend()
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

data = np.loadtxt('datasets/food_truck.csv', delimiter=',')

X = data[:, 0:1]
y = data[:, 1:2]

theta_vals = np.arange(-7.5,9.25,0.1)
J_vals = []

for theta in theta_vals:
    J = mean_squared_error(X, y, theta)
    J_vals.append(J)

plt.xlabel(r'$\theta$')
plt.ylabel(r'$J(\theta)$')
plt.plot(theta_vals, J_vals)
plt.show()
```

## Plot mean squared error for multivariable hypothesis

```python
 import numpy as np
from matplotlib import pyplot as plt

data = np.loadtxt('datasets/food_truck.csv', delimiter=',')

m = data.shape[0]
X = np.hstack((np.ones((m, 1)), data[:, 0:1]))
y = data[:, 1:2]

# Ploting cost function.

theta0_vals = np.arange(-10, 10,0.1)
theta1_vals = np.arange(-1, 4, 0.1)

J_vals = np.empty((0,len(theta0_vals)))

for theta1 in theta1_vals:
    J_vals_for_theta1 = np.array([])
    
    for theta0 in theta0_vals:
        theta = np.array([[theta0],[theta1]])
        J = mean_squared_error(X, y, theta)
        J_vals_for_theta1 = np.append(J_vals_for_theta1, J)
       
    J_vals = np.append(J_vals, np.array([J_vals_for_theta1]), axis=0)

theta0_grid, theta1_grid = np.meshgrid(theta0_vals, theta1_vals)
 
ax = plt.axes(projection='3d')
ax.set_xlabel(r'$\theta_0$')
ax.set_ylabel(r'$\theta_1$')
ax.set_zlabel(r'$J(\theta)$');
ax.plot_surface(theta0_grid, theta1_grid, J_vals, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
plt.show()

# Ploting cost function for each iteration using contour plot.

plt.contour(theta0_grid, theta1_grid, J_vals, 20)
plt.show()
```

## Calculate hypothesis for noralized features

```python
import numpy as np
from matplotlib import pyplot as plt

# Columns: sizeof the house, number of bedrooms, price.
data = np.loadtxt('datasets/house_prices.csv', delimiter=',')

m = data.shape[0]
X = data[:, 0:2]
y = data[:, 2:3]

X_norm, mu, sigma = feature_normalize(X)
X_norm = np.hstack((np.ones((m, 1)), X_norm[:, 0:2]))

# Ploting cost function for each iteration.

iterations = 800
J_hist = np.array([])
alpha = 0.01
theta = np.array([[0], [0], [0]])

for _ in range(iterations):
    J = mean_squared_error(X_norm, y, theta)
    J_hist = np.append(J_hist, J)
    theta = gradient_descent(X_norm, y, theta, alpha)

plt.xlabel('epochs')
plt.ylabel(r'$J(\theta)$')
plt.plot(J_hist)
plt.show()

# Predicting price for 1650 square feet house with 3 bedrooms.
X = np.array([[1650, 3]])
X_norm = (X - mu) / sigma
X_norm = np.append(1, X_norm)
price = X_norm.dot(theta)
print('Price:', price)
```

## Comparing gradient descent with and without learning rate decay

```python
import numpy as np
from matplotlib import pyplot as plt

# Columns: sizeof the house, number of bedrooms, price.
data = np.loadtxt('datasets/house_prices.csv', delimiter=',')

m = data.shape[0]
X = data[:, 0:2]
y = data[:, 2:3]

X_norm, mu, sigma = feature_normalize(X)
X_norm = np.hstack((np.ones((m, 1)), X_norm[:, 0:2]))

plt.figure(figsize=(16, 8))

# Ploting cost function for each iteration.
iterations = 100
alpha = 0.03
theta = np.array([[0], [0], [0]])
J_hist = np.array([])
alpha_hist = np.full((iterations, 1), alpha)

for _ in range(iterations):
    J = mean_squared_error(X_norm, y, theta)
    J_hist = np.append(J_hist, J)
    theta = gradient_descent(X_norm, y, theta, alpha)

plt.subplot(2, 2, 1)
plt.title("Constant learning rate.")
plt.xlabel('epochs')
plt.ylabel(r'$J(\theta)$')
plt.plot(J_hist)

plt.subplot(2, 2, 3)
plt.xlabel('epochs')
plt.ylabel('learning rate')
plt.plot(alpha_hist)

# Ploting cost function for each iteration using decay learning rate.
iterations = 8
eta = 0.2
decay_rate = 0.25
theta = np.array([[0], [0], [0]])
J_hist = np.array([])
eta_hist = np.array([])

for iteration in range(iterations):
    J = mean_squared_error(X_norm, y, theta)
    J_hist = np.append(J_hist, J)
    eta = decay_learning_rate(eta, decay_rate, iteration)
    eta_hist = np.append(eta_hist, eta)
    theta = gradient_descent(X_norm, y, theta, eta)
    
plt.subplot(2, 2, 2)
plt.title("Decay learning rate.")
plt.xlabel('epochs')
plt.ylabel(r'$J(\theta)$')
plt.plot(J_hist)

plt.subplot(2, 2, 4)
plt.xlabel('epochs')
plt.ylabel('learning rate')
plt.plot(eta_hist)

plt.show()
```

## Plot hypothesis using different functions

```python
import numpy as np
from matplotlib import pyplot as plt

# Columns: population of city, profit
data = np.loadtxt('datasets/food_truck.csv', delimiter=',')

m = data.shape[0]
X = data[:, 0:1]
y = data[:, 1:2]

X, _, _ = feature_normalize(X)
X = X + np.abs(np.min(X))

plt.figure(figsize=(16, 12))

# Linear function
X_linear = np.hstack((np.ones((m, 1)), X[:, 0:1]))
iterations = 800
alpha = 0.01
theta = np.array([[0], [0]])

for _ in range(iterations):
    theta = gradient_descent(X_linear, y, theta, alpha)

plt.subplot(2, 2, 1)
plt.title(r'$h(\theta)=\theta_0x_0 + \theta_1x_1$')
plt.xlabel('Population of City (normalized)')
plt.ylabel('Profit in $10,000s')
plt.plot(X, y, 'xr')

min_max = np.array([[np.min(X)], [np.max(X)]])
min_max_with_x0 = np.hstack((np.ones((2,1)), min_max))
pred = min_max_with_x0.dot(theta) 
plt.plot(min_max, pred)

# Quadratic function
X_quadratic = np.hstack((np.ones((X.shape[0], 1)), X, (X * X)))
iterations = 80
alpha = 0.01
theta = np.array([[0], [0], [0]])

for _ in range(iterations):
    theta = gradient_descent(X_quadratic, y, theta, alpha)

plt.subplot(2, 2, 2)
plt.title(r'$h(\theta)=\theta_0x_0 + \theta_1x_1 + \theta_2x_1^2$')
plt.xlabel('Population of City (normalized)')
plt.ylabel('Profit in $10,000s')
plt.plot(X, y, 'xr')

min_max = np.arange(np.min(X), np.max(X), 0.1)[:, np.newaxis]
min_max_with_x0 = np.hstack((np.ones((min_max.shape[0], 1)), min_max, (min_max*min_max)))
pred = min_max_with_x0.dot(theta) 
plt.plot(min_max, pred)

# Square root function
X_square_root = np.hstack((np.ones((X.shape[0], 1)), X, np.sqrt(X)))
iterations = 20
alpha = 0.01
theta = np.array([[0], [0], [8]])

for _ in range(iterations):
    theta = gradient_descent(X_square_root, y, theta, alpha)

plt.subplot(2, 2, 3)
plt.title(r'$h(\theta)=\theta_0x_0 + \theta_1x_1 + \theta_2\sqrt{x_1}$')
plt.xlabel('Population of City (normalized)')
plt.ylabel('Profit in $10,000s')
plt.plot(X, y, 'xr')

min_max = np.arange(np.min(X), np.max(X), 0.1)[:, np.newaxis]
min_max_with_x0 = np.hstack((np.ones((min_max.shape[0], 1)), min_max, np.sqrt(min_max)))
pred = min_max_with_x0.dot(theta) 
plt.plot(min_max, pred)

# Polynomial function
X_polynomial = np.hstack((np.ones((X.shape[0], 1)), X, (X * X), (X * X * X)))
iterations = 3000
alpha = 0.001
theta = np.array([[1], [10], [-9], [2]])

for _ in range(iterations):
    theta = gradient_descent(X_polynomial, y, theta, alpha)

plt.subplot(2, 2, 4)
plt.title(r'$h(\theta)=\theta_0x_0 + \theta_1x_1 + \theta_2x_1^2 + \theta_3x_1^3$')
plt.xlabel('Population of City (normalized)')
plt.ylabel('Profit in $10,000s')
plt.plot(X, y, 'xr')

min_max = np.arange(np.min(X), np.max(X), 0.1)[:, np.newaxis]
min_max_with_x0 = np.hstack((np.ones((min_max.shape[0], 1)), min_max, (min_max*min_max), (min_max*min_max*min_max)))
pred = min_max_with_x0.dot(theta) 
plt.plot(min_max, pred)

plt.show()
```
